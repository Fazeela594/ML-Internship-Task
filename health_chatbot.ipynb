```python
from transformers import pipeline
```

```python
chatbot = pipeline("text-generation", model="gpt2")
```

```python
def health_chatbot(question):
    prompt = f"""
    You are a helpful and friendly health assistant.
    Give simple, safe and general health information only.
    Do not give dangerous medical advice.

    Question: {question}
    Answer:
    """

    response = chatbot(prompt, max_length=100, num_return_sequences=1)
    return response[0]['generated_text']
```

```python
print(health_chatbot("What causes fever?"))
```

```python
print(health_chatbot("How to reduce stress?"))
```
